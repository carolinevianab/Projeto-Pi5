```python
import gym
import random
from nes_py.wrappers import JoypadSpace
from Contra.actions import SIMPLE_MOVEMENT, COMPLEX_MOVEMENT, RIGHT_ONLY
```

```python
#https://pypi.org/project/gym-contra/

env = gym.make('Contra-v0')

# informações de cada 'package' de ações -> https://github.com/OuYanghaoyue/gym_contra/blob/master/Contra/actions.py
#RIGHT_ONLY, SIMPLE_MOVEMENT, COMPLEX_MOVEMENT
package = RIGHT_ONLY
env = JoypadSpace(env, package)

print("Observation Space: ",env.observation_space)
print("Actions: ", env.action_space)
```

```python
class Agent():
    def __init__(self, env):
        self.action_size = env.action_space.n
    
    def get_action(self):
        action = random.choice(range(self.action_size))
        return action
```

```python
agent = Agent(env)

done = False
env.reset()

while not done:
    action = agent.get_action()
    state, reward, done, info = env.step(action)
    #print(action)
    env.render()

print("Final Score: ", reward)
```
